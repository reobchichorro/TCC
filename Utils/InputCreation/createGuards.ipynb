```python
import numpy as np
import os
import pandas as pd
df = pd.read_csv(filepath_or_buffer='GuardDatabase/icosts.csv', sep='\t', lineterminator='\n', header=0)
```

```python
path = '../../Inputs/GuardTypes/'
for k in df.itertuples():
    name = 'h' + str(k.height) + '-r' + str(k.radius) + '-a' + str(k.angle)
    # print(name)
    with open(path + name, 'w') as f:
        f.write(name+'\n')
        f.write(str(k.height)+'\n')
        f.write(str(k.radius)+'\n')
        f.write(str(k.angle)+'\n')
        f.write(str(k.per100000Covered)+'\n')
        f.write('0\n')
        f.write('10\n')
    # break
```

```python
lines = []
for k in df.itertuples():
    name = 'h' + str(k.height) + '-r' + str(k.radius) + '-a' + str(k.angle)
    lines.append((name, k.height, k.radius, k.angle, k.per100000Covered, 0, 10))
print(lines)
```

```python
import random

guardTypeList = '../../Inputs/GuardTypeList/'

for i in range(0,100):
    numGuards = random.randint(1,8)
    guards = random.sample(lines, numGuards)
    with open(guardTypeList + str(i).zfill(4), 'w') as f:
        f.write('/home/reobc/Documents/Disciplinas/TCC/Inputs/GuardTypes/ ' + str(numGuards) + '\n')
        for j in range(0,numGuards):
            f.write(str(guards[j][0])+'\n')


# /home/reobc/Documents/Disciplinas/TCC/Inputs/GuardTypes/ 2
# 0000
# 0001
```
